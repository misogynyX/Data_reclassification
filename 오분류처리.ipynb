### 재분류가 필요한 idx는 다음과 같이 리스트 형태로 저장해 주세요

```
def corr_idx (error_idx,df):
  """idx를 이용하여 잘못 구분된 라벨 수정하는 코드"""

  error_id = list(sorted(set(error_idx)))
  df.loc[error_idx,'label'] = 0
```
