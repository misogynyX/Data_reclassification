```
def txt_preprocessing(sample):
    sample.drop(['article_id','cp_name','authors','keywords','date','url','tags'],axis = 1, inplace = True)
    sample['text'] = sample['text'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣0-9a-zA-Z ]","")
    sample['text'] = sample['text'].str.lower()
    
    return sample
```
